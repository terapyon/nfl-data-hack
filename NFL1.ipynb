```python
g3= d1.query("Quarter==1")[["Minute", "Second", "OffenseTeam", "YardLine","PlayType"]]
```

```python
def left_to(team, yard):
    if team == "DEN":
        return 100 - yard
    else:
        return yard
```

```python
def to_sec(m, s):
    return m + s/60
```

```python
g3["m"] = g3.apply(lambda x: to_sec(x.Minute, x.Second), axis=1)
```

### 

```python
g3["on"] = g3.apply(lambda x: left_to(x.OffenseTeam, x.YardLine), axis=1)
```

```python
g4 = g3.sort_values(by=["Minute", "Second"], ascending=[0, 0])
g4.plot.scatter(x="on", y="m")
```

```python
d1 = dataset.query("OffenseTeam=='DEN'|DefenseTeam=='DEN'").query("GameDate=='2016-01-03'")
```

```python
def plot_quarter(q):
    df= d1.query("Quarter=="+str(q))[["Minute", "Second", "OffenseTeam", "YardLine","PlayType"]]
    df["m"] = df.apply(lambda x: to_sec(x.Minute, x.Second), axis=1)
    df["on"] = df.apply(lambda x: left_to(x.OffenseTeam, x.YardLine), axis=1)
    df = df.sort_values(by=["Minute", "Second"], ascending=[0, 0])
    df.plot.scatter(x="on", y="m")
```

```python
plot_quarter(1)
```

```python
plot_quarter(2)
```

```python
plot_quarter(3)
```

```python
plot_quarter(4)
```
